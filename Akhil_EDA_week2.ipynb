```python
Popular_director = df.groupby("director")["popularity"].mean().sort_values(ascending=False).head(5)
print(Popular_director)
```

```python
Leastpopular_director = df.groupby("director")["popularity"].mean().sort_values(ascending=False).tail(5)
print(Leastpopular_director)
```

```python
Popular_genres = df.groupby("genres")["popularity"].mean().sort_values(ascending=False).head(10)
print(Popular_genres)
```

```python
Leastpopular_genres = df.groupby("genres")["popularity"].mean().sort_values(ascending=False).tail(10)
print(Leastpopular_genres)
```

```python
AverageBudgetbygenres = df.groupby("genres")["budget"].mean().sort_values(ascending=False).head(10)
print(AverageBudgetbygenres)
```

```python
Budgetbygenres = df.groupby("genres")["budget"].sum().sort_values(ascending=False).head(10)
print(Budgetbygenres)
```

```python
Revenuebydirector = df.groupby("director")["revenue"].mean().sort_values(ascending=False)
print(Revenuebydirector)
```

```python
Revenuebytitle = df.groupby("title")["revenue"].sum().sort_values(ascending=False)
print(Revenuebytitle)
```

```python
Budgetbytitle = df.groupby("title")["budget"].sum().sort_values(ascending=False)
print(Budgetbytitle)
```

```python

plt.title("Scatter Plot for Revenue and Budget")
plt.scatter(df['revenue'], df['budget'], color='red')
plt.xlabel("Budget")
plt.ylabel("Revenue")
plt.show()
```

```python
plt.title("Genres Comparison")
plt.bar(Popular_genres.index, Popular_genres.values, color='orange')
plt.xlabel("Genres")
plt.ylabel("Average Popularity")
plt.xticks(rotation=90)
plt.show()
```

```python
genre_trend = df.groupby('genres')['vote_count'].mean().head(10)
print(genre_trend)
```

```python
plt.title("Genres Trend Analysis")
plt.plot(genre_trend.index,genre_trend.values, marker='o', color='green')
plt.xlabel("Genres")
plt.ylabel("Average_votes")
plt.xticks(rotation=90)
plt.figure(figsize=(16, 12))
plt.show()
```

```python
plt.title('Genres Budget')
sns.boxplot(x=AverageBudgetbygenres.index, y=AverageBudgetbygenres.values, color="#FFD700", linewidth=1.5)
plt.xlabel('Genre Name')
plt.ylabel('Genre Average Budget')
plt.xticks(rotation=90)
plt.show()
```
